# **A transformation optics approach to singular metasurfaces**

**Abstract**:
Recent advancements propose geometrical singularities in plasmonic metasurfaces to enhance light-matter interactions, exploiting their broad light-harvesting capabilities and extreme plasmon confinement. Metal surfaces' optical response is predominantly governed by surface plasmons, intricately tied to surface geometry. This study delves into metasurfaces harboring singularities manifested as sharp edges, exhibiting three quantum numbers despite their two-dimensional surface. Our investigation encompasses the characterization of plasmonic excitations, their capacity for high optical energy concentrations, and the transition from discrete excitation spectra of non-singular surfaces to the continuous spectra of singular metasurfaces. Specifically, a metallic metasurface is examined, periodically structured as a grating with hidden compacted dimensions, featuring a series of singularities. These singularities introduce an additional dimension, altering the nature of surface plasmon excitations. Unlike typical two-dimensional metal gratings characterized by two wave vectors, plasmonic excitations in the studied metasurfaces are characterized by three wave vectors.

## 1. Introduction
### **Plots have been reproduced Section 3 onwards
A conventional two-dimensional (2D) object is characterized by two quantum numbers. For example, the frequencies of surface plasmons on a periodic surface are labeled by the components of their momentum projected onto the surface axes. We describe theoretically systems that instead require three quantum numbers to label them: the two conventional in-plane momenta plus a third momentum corresponding to a compacted dimension hidden from view inside a singularity.

We make use of the technique of transformation optics, which exploits the invariance of Maxwell's equations under a coordinate transformation: Only the values of $\epsilon$, $\mu$ are affected by the transformation. We used this theory to compact a dimension through a singular transformation that compresses one of the dimensions of a 3D system into one or more singular points.

Singular metasurfaces constitute a special class of structured surfaces which feature sharp edges or ultra-small gaps. A conventional metasurface is characterized by two wave vectors, such that the selection of k-vectors is discretized. On the other hand, singular metasurfaces feature three wave vectors, owing to the additional length scale introduced by the singularity, which allows surface modes to exist over a continuum of quantum numbers. However, while a local singular structure has a broadband optical response, nonlocality sets a length scale that prevents the formation of a perfect singularity, thus yielding a discrete resonance spectrum.

The singular metasurface has some unusual properties. The modes of periodic array of metal slabs are truly 3D, being spread out over the whole structure, and are characterized by the three wave vectors, $k_x$, $k_y$, $k_z$. By contrast, these modes when transformed are now found to be exponentially localized at the interface and at first sight would seem to be 2D objects. However, this is not the case; the structure inherits the spectral characterization of the mother structure and therefore is labeled by the same three wave vectors, $k_x$, $k_y$, $k_z$. This affects the response of the metasurface to external stimulae. A conventional grating, free of any singularities such as sharp edges, is characterized by only two wave vectors, which are defined by the angle of incidence of external radiation. This leads to a discrete excitation spectrum. In other words, ordinary gratings appear colored to the eye. By contrast, external radiation incident on our metasurface defines only two of the three wave vectors, the third being selected by the frequency. Thus, the modes form a continuum and can be excited whatever the incident angle or frequency.

## 2. Theoretical Background

### 2.1 Theory of Surface plasmons

A plasmon is a collective oscillation of the free electrons in a noble metal. It can be described as a quantum of plasma oscillations. The free electrons of a metal are treated as an electron liquid of high density of about $10^{23} cm^{-3}$ , ignoring the lattice in a first approximation. One can think of these plasma oscillations as mechanical
oscillations of the electron gas of a metal, the presence of an external electric field causing displacements of the electron gas with respect to the fixed ionic cores. For bulk plasmons (Here, bulk refers to materials with an extent large compared to the wavelength of light in all three dimensions), these oscillations occur at the plasma
frequency $\omega_p$ and have energy $E_p = \hbar \omega_p =\hbar \sqrt{\frac{ne^2}{m_0\epsilon_0}}$, where $\hbar$ denotes the Dirac constant, $n$ is the electron density, $\epsilon_0$ is the permittivity
of free space, $e$ is the electron charge, and $m_0$ is the electron mass.

### 2.2 Surface Plasmon polaritons

The key component that supports surface plasmons is metal, where the electromagnetic (EM) surface waves arise via the coupling of the EM fields to oscillations of the metal's electron plasma.

### 2.2.1 Fundamentals
* Visible light cannot propagate through highly conductive media such as metals

* When a light beam crosses the boundary into such a medium, its intensity rapidly diminishes within a short distance known as the penetration depth, which can be substantially smaller than a wavelength.

* A metallic surface acts rather like a simple mirror from which light is fully reflected back into the contiguous dielectric medium whence it came.

* However, metals can support light waves, provided that they travel along the boundaries of the metal, in regions confined to sub-wavelength dimensions. (An incident wave gets reflected from a metal surface, but in certain conditions it is able to excite some kind of surface waves on the surface of the metal)


* Such light waves travel along a metal surface in the form of a guided surface wave. It may propagate on, but not in, metal wires, and it may be guided by subwavelength metallic structures configured as integrated-photonic circuits. (useful in integrated-photonic circuits as wave-guides, can transfer information across the metallic surface over very very short distances)

* Such structures can also serve as resonators within which light may be confined, or from which light can scatter strongly at specific resonance frequencies.

* These electromagnetic waves that travel along a metal-dielectric or metal-air interface, practically in the infrared or visible-frequency are called surface plasmon polaritons.

* More specifically, surface plasmon polaritons are electromagnetic excitations propagating at the interface between a dielectric and a conductor, evanescently confined in the perpendicular direction. (They decay along the perpendicular direction, decays very sharply along the metal)



        Fig: Illustrations of surface plasmon (in the planar metal film)


* These electromagnetic surface waves arise via the coupling of the electromagnetic fields to oscillations of the conductor's electron plasma.

### 2.2.2 Wave Equation

* In order to investigate the physical properties of surface plasmon polaritons (SPPs), we have to apply Maxwell's equations to the flat interface between a conductor and a dielectric. (These are electromagnetic waves so they are bound to follow the Maxwell's equations)

* Remember, the Maxwell's curl equations:
    \begin{equation}
        \nabla \times \textbf{E} = - \frac{\partial \textbf{B}}{\partial t} \; \; \; \; \; \; \; \;  \; ---(1)
    \end{equation}
    \begin{equation}
        \nabla \times \textbf{H} = \textbf{J}_{ext} + \frac{\partial \textbf{D}}{\partial t}   \; \; \; \; \; \; \; \;  \; ---(2)
    \end{equation}

* In the absence of external charge and current densities, the Maxwell equations can be combined to yield equation:
    \begin{equation}
        \nabla \times \nabla \times \textbf{E} = - \mu_0 \frac{\partial^2 \textbf{D}}{\partial t^2}  \; \; \; \; \; \; \; \;  \; ---(3)
    \end{equation}

* Using the identities $\nabla \times \nabla \times \textbf{E}\equiv \nabla (\nabla \cdot \textbf{E})-\nabla^2\textbf{E}$ as well as $\nabla \cdot (\epsilon \textbf{E}) \equiv \textbf{E}\cdot \nabla \epsilon + \epsilon \nabla \cdot \textbf{E}$, and remembering that due to the absence of external stimuli $\nabla \cdot \textbf{D}=0$, eq(6) can be rewritten as: \begin{equation}
        \nabla \left(- \frac{1}{\epsilon}\textbf{E}\cdot \nabla \epsilon \right) - \nabla^2\textbf{E} = - \mu_0 \epsilon_0 \epsilon \frac{\partial^2 \textbf{E}}{\partial t^2}  \; \; \; \; \; \; \; \;  \; ---(4)
    \end{equation}

* For negligible variation of the dielectric profile $\epsilon = \epsilon(\textbf{r})$ over distances of the order of one optical wavelength, eq(4) simplifies to the central equation of electromagnetic wave theory:
    \begin{equation}
        \nabla^2\textbf{E} -  \frac{\epsilon}{c^2} \frac{\partial^2 \textbf{E}}{\partial t^2} = 0  \; \; \; \; \; \; \; \;  \; ---(5)
    \end{equation}

* Practically, this equation has to be solved separately in regions of constant $\epsilon$, and the obtained solutions have to be matched using appropriate boundary conditions.

* Equation (4) needs to be casted in a form suitable for the description of confined propagating waves.

* A harmonic times dependence $\textbf{E}(r,t)=\textbf{E}(r)e^{-\iota \omega t}$ of the electric field is assumed and inserted into eq (4), which yields: \begin{equation}
        \nabla^2 \textbf{E}+ k_0^2 \epsilon \textbf{E} =0  \; \; \; \; \; \; \; \;  \; ---(6)
    \end{equation}
    
    where $k_0=\omega/c$ is the wave vector of the propagating wave in vacuum.

* Equation (5) is known as the Helmholtz equation.

* Next the propagation geometry needs to be defined.




  Fig: Definition of a planar waveguide geometry. The waves propagate along the x-direction in a Cartesian coordinate system

* Assume for simplicity a one-dimensional problem, i.e. $\epsilon$ depends only on one spatial coordinate.

* Specifically, the waves propagate along the x-direction of a cartesian coordinate system, and show no spatial variation in the perpendicular, in-plane y-direction; therefore $\epsilon = \epsilon(z)$.

* Applied to electromagnetic surface problems, the plane $z=0$ coincides with the interface sustaining the propagating waves, which can now be described as $\textbf{E}(x,y,z)=\textbf{E}(z)e^{\iota \beta x}$.

* The complex parameter $\beta = k_x$ is called the propagation constant of the travelling waves and corresponds to the component of the wave-vector in the direction of propagation. (waves propagate along the x direction)

* Inserting this expression into eq(6) yields the desired form of the wave equation:
    \begin{equation}
        \frac{\partial^2 \textbf{E}(z)}{\partial z^2} + (k_0^2\epsilon - \beta^2)\textbf{E} =0  \; \; \; \; \; \; \; \;  \; ---(7)
    \end{equation}

* Naturally, a similar equation exists for the magnetic field **H**.

* Equation (7) is the starting point for the general analysis of guided electromagnetic modes in waveguides.

* In order to use the wave equation for determining the spatial field profile and dispersion of propagating waves, explicit expressions for the different field components of $\textbf{E}$ and $\textbf{H}$ need to be derived.

* This can be achieved in a straightforward way using the curl equations (1) and (2).

* For harmonic time dependence $\partial/\partial t = -\iota \omega$, we arrive at the following set of coupled equation:
    \begin{equation}
        \frac{\partial E_z}{\partial y} - \frac{\partial E_y}{\partial z} = \iota \omega \mu_0 H_x  \; \; \; \; \; \; \; \;  \; ---(8)
    \end{equation}
    \begin{equation}
        \frac{\partial E_x}{\partial z} - \frac{\partial E_z}{\partial x} = \iota \omega \mu_0 H_y   \; \; \; \; \; \; \; \;  \; ---(9)
    \end{equation}
    \begin{equation}
        \frac{\partial E_y}{\partial x} - \frac{\partial E_x}{\partial y} = \iota \omega \mu_0 H_z \; \; \; \; \; \; \; \;  \; ---(10)
    \end{equation}
    \begin{equation}
        \frac{\partial H_z}{\partial y} - \frac{\partial H_y}{\partial z} = \iota \omega \mu_0 H_x   \; \; \; \; \; \; \; \;  \; ---(11)
    \end{equation}
    \begin{equation}
        \frac{\partial H_x}{\partial z} - \frac{\partial H_z}{\partial x} = \iota \omega \mu_0 E_y  \; \; \; \; \; \; \; \;  \; ---(12)
    \end{equation}
    \begin{equation}
        \frac{\partial H_y}{\partial x} - \frac{\partial H_x}{\partial y} = \iota \omega \mu_0 E_z  \; \; \; \; \; \; \; \;  \; ---(13)
    \end{equation}

* For propagation along the x-direction $\partial/\partial x = \iota \beta$ and homogeneity in the y-direction $\partial/\partial y = 0$, this system of equation simplifies to:
    \begin{equation}
        \frac{\partial E_y}{\partial z} = -\iota \omega \mu_0 H_x  \; \; \; \; \; \; \; \;  \; ---(14)
    \end{equation}
    \begin{equation}
        \frac{\partial E_x}{\partial z} - \iota \beta E_z = \iota \omega \mu_0 H_y  \; \; \; \; \; \; \; \;  \; ---(15)
    \end{equation}
    \begin{equation}
        \iota \beta E_y = \iota \omega \mu_0 H_z \; \; \; \; \; \; \; \;  \; ---(16)
    \end{equation}
    \begin{equation}
        \frac{\partial H_y}{\partial z} = \iota \omega \epsilon_0 \epsilon E_x \; \; \; \; \; \; \; \;  \; ---(17)
    \end{equation}
    \begin{equation}
        \frac{\partial H_x}{\partial z} - \iota \beta H_z = -\iota \omega \epsilon_0\epsilon E_y  \; \; \; \; \; \; \; \;  \; ---(18)
    \end{equation}
    \begin{equation}
        \iota \beta H_y = -\iota \omega \epsilon_0 \epsilon E_z  \; \; \; \; \; \; \; \;  \; ---(19)
    \end{equation}

* It can easily be shown that this system allows two sets of self-consistent solutions with different polarization properties of the propagating waves.

* The first set are the transverse magnetic (TM or p) modes, where only the field components $E_x$, $E_z$ and $H_y$ are nonzero, and the second set the transverse electric (TE or s) modes, with only $H_x$, $H_z$ and $E_y$ being nonzero.

* For TM modes, the system of governing equations (17-22) reduces to:
    \begin{equation}
        E_x = -\iota \frac{1}{\omega \epsilon_0\epsilon}\frac{\partial H_y}{\partial z}  \; \; \; \; \; \; \; \;  \; ---(20)
    \end{equation}
    \begin{equation}
        E_z = -\frac{\beta}{\omega \epsilon_0\epsilon} H_y \; \; \; \; \; \; \; \;  \; ---(21)
    \end{equation} and the wave equation for TM modes is:
    \begin{equation}
        \frac{\partial^2 H_y}{\partial z^2} + (k_0^2 -\beta^2)H_y = 0 \; \; \; \; \; \; \; \;  \; ---(22)
    \end{equation}

* For TE modes the analogous set is:
    \begin{equation}
        H_x = \iota \frac{1}{\omega \mu_0} \frac{\partial E_y}{\partial z} \; \; \; \; \; \; \; \;  \; ---(23)
    \end{equation}
    \begin{equation}
        H_z = \frac{\beta}{\omega \mu_0} E_y \; \; \; \; \; \; \; \;  \; ---(24)
    \end{equation} with the TE wave equation
    \begin{equation}
        \frac{\partial^2 E_y}{\partial z^2} + (k_0^2 \epsilon - \beta^2)E_y = 0 \; \; \; \; \; \; \; \;  \; ---(25)
    \end{equation}

* With these equations, surface plasmon polaritons can be investigated mathematically.

### 2.2.3 SPPs at a single interface

* The most simple geometry sustaining SPPs is that of a single, flat interface between a dielectric, non-absorbing half space ($z>0$) with positive real dielectric constant $\epsilon_2$ and an adjacent conducting half space ($z<0$) described via a dielectric function $\epsilon_1(\omega)$ (complex).



 Fig: Geometry for SPP propagation at a single interface between a metal and a dielectric

* The propagating wave solutions confined to the interface, i.e. with evanescent decay in the perpendicular z-direction can be obtained using equation set (20-22; i.e., TM modes) in both half-spaces

   1. for $z>0$
        \begin{equation}
            H_y(z) = A_2 e^{\iota \beta x}e^{-k_2 z} \; \; \; \; \; \; \; \;  \; ---(26)
        \end{equation}
        \begin{equation}
            E_x(z) = \iota A_2 \frac{1}{\omega \epsilon_0 \epsilon_2}k_2 e^{\iota \beta x}e^{-k_2 z} \; \; \; \; \; \; \; \;  \; ---(27)
        \end{equation}
        \begin{equation}
            E_z(Z) = -A_1 \frac{\beta}{\omega \epsilon_0 \epsilon_2}e^{\iota \beta x}e^{-k_2 z} \; \; \; \; \; \; \; \;  \; ---(28)
        \end{equation}
        
  2. for $z<0$ (replace z with -z in above set of equations)
        \begin{equation}
            H_y(z) = A_1 e^{\iota \beta x}e^{k_1 z} \; \; \; \; \; \; \; \;  \; ---(29)
        \end{equation}
        \begin{equation}
            E_x(z) = -\iota A_1 \frac{1}{\omega \epsilon_0 \epsilon_1}k_1 e^{\iota \beta x}e^{k_1 z} \; \; \; \; \; \; \; \;  \; ---(30)
        \end{equation}
        \begin{equation}
            E_z(Z) = -A_1 \frac{\beta}{\omega \epsilon_0 \epsilon_1}e^{\iota \beta x}e^{k_1 z} \; \; \; \; \; \; \; \;  \; ---(31)
        \end{equation}

* $k_i\equiv k_{z,i}$ (i=1,2) is the component of the wave vector perpendicular to the interface in the two media.

* Its reciprocal value, $\hat{z}=1/|k_z|$, defines the evanescent decay length of the fields perpendicular to the interface, which qualifies the confinement of the wave (along each side).

* Continuity of $H_y$ and $\epsilon_iE_z$ at the interface requires that $A_1=A_2$ and: \begin{equation}\frac{k_2}{k_1}=-\frac{\epsilon_2}{\epsilon_1}, \; \; \; \; \; \; \; \;  \; ---(32) \end{equation} that is the two permittivities $\epsilon_1$ and $\epsilon_2$ must be opposite in signs.

* Note that with our convention of the signs in the exponents in (26-31), confinement to surface demands $Re[\epsilon_1]<0$ if $\epsilon_2>0$ - the surface waves exist only at interfaces between materials with opposite signs of the real part of their dielectric permittivities; possible only when you choose one as metal and other as dielectric (that is metal-dielectric interface will actually be able to support surface plasmon waves).
    
* The expression for $H_y$ further has to fulfill the wave equation (22), yielding:
    \begin{equation}
        k_1^2 = \beta^2 - k_0^2 \epsilon_1 \; \; \; \; \; \; \; \;  \; ---(33)
    \end{equation}
    \begin{equation}
        k_2^2 = \beta^2 - k_0^2 \epsilon_2 \; \; \; \; \; \; \; \;  \; ---(34)
    \end{equation} where $k_1$ and $k_2$ is the perpendicular component of the wave vector along z-direction into the metal and dielectric resp.

* Commbining equations (32-34), the **dispersion relation of SPPs** propagating at the interface between the two half spaces:
    \begin{equation}
        \beta = k_0 \sqrt{\frac{\epsilon_1\epsilon_2}{\epsilon_1+\epsilon_2}} \; \; \; \; \; \; \; \;  \; ---(35)
    \end{equation} This expression is valid for both real and complex $\epsilon_1$, i.e., for conductors without and with attenuation.
    
* Before discussing the properties of the dispersion relation (35) in more detail, the possibility of TE surface modes are briefly analyzed.

* Using (20-35), the respective expressions for the field components are:

    1. for $z>0$

        \begin{equation}
            E_y(z) = A_2 e^{\iota \beta x} e^{-k_2z} \; \; \; \; \; \; \; \;  \; ---(36)
        \end{equation}
        \begin{equation}
            H_x(z) = -\iota A_2 \frac{1}{\omega \mu_0}k_2 e^{\iota \beta x} e^{-k_2z} \; \; \; \; \; \; \; \;  \; ---(37)
        \end{equation}
        \begin{equation}
            H_z(z) = A_2 \frac{\beta}{\omega \mu_0} e^{\iota \beta x} e^{-k_2z} \; \; \; \; \; \; \; \;  \; ---(38)
        \end{equation}
    2. for $z<0$

        \begin{equation}
            E_y(z) = A_1 e^{\iota \beta x} e^{k_1z} \; \; \; \; \; \; \; \;  \; ---(39)
        \end{equation}
        \begin{equation}
            H_x(z) = \iota A_1 \frac{1}{\omega \mu_0}k_1 e^{\iota \beta x} e^{k_1z} \; \; \; \; \; \; \; \;  \; ---(40)
        \end{equation} 
        \begin{equation}
            H_z(z) = A_1 \frac{\beta}{\omega \mu_0} e^{\iota \beta x} e^{k_1z} \; \; \; \; \; \; \; \;  \; ---(41)
        \end{equation}

* Continuity of $E_y$ and $H_x$ at the interface leads to the condition:
    \begin{equation}
        A_1(k_1+k_2)=0 \; \; \; \; \; \; \; \;  \; ---(42)
    \end{equation}
    
* Since confinement to the surface requires $Re[k_1]>0$ and $Re[k_2]>0$, this condition is only fulfilled if $A_1 =0$, so that also $A_2 =A_1=0$.
    
* Thus, **no surface modes exist for TE polarization**.

* **Surface plasmon polaritons only exist for TM polarization**.

### 2.2.5 Generation of SPPs

* Surface plasmon polaritons on a flat metal/dielectric **interface cannot be excited directly by light beams since $\beta>k$**, where $k$ is the wavevector of light on the dielectric side of the interface and $\beta$ is the propagation constant of SPPs. (At a particular frequency momentum of plasmons is higher than that of photons, hence because of this momentum mismatch energy transfer isnt possible from photons to plasmons)

* Therefore, **the projection along the interface of the momentum** $k_x = k \sin{\theta}$ of photons impinging under an angle $\theta$ to the surface normal **is always smaller than the SPP propagation constant $\beta$**, even at grazing incidence, prohibiting phase-matching.

* However, phase-matching to SPPs can be achieved in a **three-layer system** consisting of a *thin metal film sandwiched between two insulators* of different dielectric constants: **Kretschmann \& Otto configuration** <br>


**Prism Coupling**



 Fig: Prism coupling to SPPs using attenuated total internal reflection in the Kretschmann (left) and Otto (right) configuration. Also drawn are possible light paths for excitation



* For simplicity, we will take one of the insulators to be air ($\epsilon$=1). A beam reflected at the interface between the insulator of higher dielectric constant $\epsilon$, usually in the form of a prism.
* The most common configuration is the **Kretschmann** method, in which a thin metal film is evaporated on top of a glass prism.
* Photons from a beam impinging from the glass side at an angle greater than the critical angle of total internal reflection tunnel through the metal film and excite SPPs at the metal/air interface.

* The metal will have an in-plane momentum $k_x = k\sqrt{\epsilon}\sin{\theta}$, which is sufficient to excite SPPs at the interface between the metal and the lower-index dielectric, i.e., in this case at the metal/air interface. (From the below plot, we see that there is a possibility that plasmons can have lower momentum than the incoming photons (light cone of prism) on metal-air interface for a given frequency; so by changing the angle of incidence, we will be able to match the $k_x$ component with the $\beta$ of surface plasmon; hence exciting SPPs on metal-air interface is possible but not on metal-prism interface)



 Fig: Prism coupling and SPP dispersion. Only propagation constants between the light lines of air and the prism (usually glass) are accessible, resulting in additional SPP damping due to leakage radiation into the latter: the excited SPPs have propagation constants inside the prism light cone


* This way, SPPs with propagation constants $\beta$ between the light lines of air and the higher-index dielectric (here, prism) can be excited.

* Note that phase-matching to SPPs at the prism/metal interface cannot be achieved, since the respective SPP dispersion lies outside the prism light cone.

* This **prism coupling scheme** - known as *attenuated total internal reflection* - involves tunneling of the fields of the excitation beam to the metal/air interface where SPP excitation takes place.

* In short, the SPPs are excited using phase-matching via $\beta = k\sqrt{\epsilon}\sin{\theta}$

* Another geometry is the **Otto configuration** in which the prism is separated from the metal film by a thin air gap.

* Total internal reflection takes place at the prism/air interface, exciting SPPs via tunneling to the air/metal interface.

* This configuration is preferable when direct contact with the metal surface is undesirable, for example for studies of surface quality.


**Grating Coupling**




Fig: Phase-matching of light to SPPs using a grating

* The mismatch in wave vector between the in-plane moment $k_{||}=k\sin{\theta}$ of impinging photons and $\beta$ can also be overcome by patterning the metal surface with a shallow grating of grooves or holes with lattice constant $a$.

* For the simple one-dimensional grating, phase matching takes place whenever the condition is fulfilled: $\beta = k\sin{\theta}\pm \nu g$ where $g = 2\pi/a$ is the reciprocal vector of the grating, and $\nu = (1,2,3,...)$.

* Physical Intuition: Because of this grating when light falls on it, it will allow you to excite all different modes not only the fundamental mode, so there is a possibility that with one of these modes, the propagation constant of plasmons $\beta$ will have a phase match and hence will be able to excite.

### 2.2.6 Excitation of surface plasmon polaritons

Surface plasmons can be excited by both electrons and photons. The excitation by electrons is created by firing electrons into the bulk of a metal. As the electrons scatter, energy is transferred into the bulk plasma. The component of the scattering vector parallel to the surface results in the formation of surface plasmons. For photon excitations, SPPs on a flat metal/insulator interface cannot be excited directly by light beams since the surface plasmon propagation constant is larger than the free space wave vector (i.e. $k_x > k_0$). Therefore, certain techniques that can achieve phase-matching or spatial mode matching are critical.

### 2.3 Metasurfaces

In conventional materials $\epsilon$, $\mu$ derive from the constituent atoms. While in metamaterials $\epsilon_{eff}$ , $\mu_{eff}$ derive from the sub-units which may contain many atoms.

The response of conventional materials to
electromagnetic fields is determined by their constituent molecules. In contrast metamaterials owe their properties to much larger units, though still less than the wavelength of radiation. These large structural units, or metamolecules can be designed to give a much wider spectrum of properties.

Metamaterials offer a huge range of new electromagnetic properties: negative refraction, spatial inhomogeneity to name only two. To exploit the possibilities offered in this new world we need a new design tool. 

Maxwells equations are exact at the classical level but lack transparency; Snells law is elegantly visual, an aid to the imagination, but fails to account for many vital aspects of electromagnetism.

Transformation optics retains an intuitive appeal, replacing the rays of Snells law with the field lines of Maxwell whose equations represents exactly.


**Why we need Metamaterials**

* Modern material engineering leads to the synthesis of many novel chemical compounds by combining atoms of natural materials to obtain some desired electromagnetic property

* Nanoscale devices made up of these novel compounds can exhibit the combined or modified electromagnetic response which cannot be obtained individually from the constituent materials

* However, the major bottleneck of this material engineering for nanoscale devices is the limited set of natural materials appearing in periodic table exhibiting a fixed set of electromagnetic properties

### 2.3.1 Meta-atoms

### 2.3.2 Singular Metasurfaces

From traditional concepts, it is usually believed that a metallic structure should
have a large physical size (as compared to the wavelength) to allow for a broadband light-harvesting process and a nanoparticle of finite size usually sustains localized surface plasmon resonances at discrete, rather than continuous frequencies. However,
there are exceptions to these rules. Some finite nanostructures containing sharp edges (or corners) can behave like infinite plasmonic systems, and show a continuous interaction with light over a broad frequency range.

The theory of TO enables the understanding of the physical mechanism behind this phenomenon, which suggests that the sharp geometrical features in these nanostructures act as singularities
for surface plasmons, causing them to propagate towards the sharp points, slowing down as they progress, but never reaching these sharp points. Consequently, light energy builds up around the singularities, and the resonance spectrum becomes continuous.

### 2.4 Transformation Optics

### 2.4.1 Fundamentals
* Transformation Optics is a mathematical tool that facilitates the design of optical materials that guide light along desired trajectories

* The underlying concept relies on a geometrical transformation that converts simple trajectories into desired ones.

* In order that Maxwell's equations remain valid, the optical parameters associated with the transformed equivalent system must also be modified, and this establishes the character of the required optical material.

* As a simple example of such equivalence, **a local compression of the coordinate system by a scaling factor is equivalent to a local increase of the refractive index by the same factor**, so that the **optical path length** (product of length and refractive index) **remains unchanged**.
  
* A three-step design procedure provides a guide:
  1. begin with a *pilot* physical system for which the optical trajectories are known, such as homogeneous and isotropic material.
  2.  Find a coordinate transformation that converts these trajectories to the desired ones.
  3. Determine the transformed physical parameters of the equivalent material.
  4. The new material will implement the desired optical trajectories in the original coordinate system
  
* Since geometrical transformations generally involve changes of directions and introduce direction-dependent scaling, the transformed parameters are generally both anisotropic and spatially varying.







Fig: Geometrical transformation implementing refraction without reflection. (a) Desired optical trajectories. (b) Free space with straight-line optical trajectories. (c) Stretching the coordinate system by the factor s for x3 > 0 causes the rays to change slope and follow the desired trajectories. (d) Equivalent anisotropic, homogeneous material that causes the rays to change slope in an identical way.

### 2.4.2 Principle
* Let $\{\epsilon_{ij}\}$ and $\{\mu_{ij}\}$ be the elements of the permittivity and permeability tensors of the original material in the original coordinate system $(x_1, x_2, x_3)$.

* The elements of the permittivity and permeability tensors of the equivalent material (denoted by the superscript "e") in the transformed coordinate system $(u_1, u_2, u_3)$ are then related to the original elements by the matrix equations:
    \begin{equation}
        \epsilon^{e} = \frac{\textbf{A}^T \epsilon \textbf{A}}{|det(\textbf{A})|}
    \end{equation}
    \begin{equation}
        \mu^e = \frac{\textbf{A}^T \mu \textbf{A}}{|det(\textbf{A})|}
    \end{equation}

* Here $\textbf{A}$ is the $3\times 3$ Jacobian transformation matrix, whose elements are the partial derivatives:
    \begin{equation}
        \textbf{A}_{ij} = \frac{\partial u_i}{\partial x_j} \; \; \; \; \; \; \; \; \; \; \; \; \; i,j=1,2,3....
    \end{equation}
    Since **A** is generally dependent on $(x_1, x_2, x_3)$, the equivalent material is generally inhomogeneous, even if the original material is homogeneous.    

* In the special case for which the original material is both homogeneous and isotropic, say free space, then $\epsilon$ and $\mu$ are diagonal with equal diagonal elements $\epsilon_0$ and $\mu_0$, respectively, whereupon:
    \begin{equation}
        \epsilon_0^{-1}\epsilon^e = \mu_0^{-1}\mu^e = |det(\textbf{A})|^{-1}\textbf{A}^T \textbf{A}
    \end{equation}

    The tensors $\epsilon^e$ and $\mu^e$ are then identical except for a scaling factor ($\epsilon_0^{-1}$ and $\mu_0^{-1}$ resp.).Under these conditions the impedance, which depends on their ratio, remains unchanged for all polarizations, which in turn implies that the equivalent medium introduces no reflection at any boundary with free space. (hence, becomes a reflection free boundary) 

### 2.4.3 Conformal Transformation

A conformal mapping is an analytic transformation that preserves local angles. If we consider an analytic function $\phi(z)$ in the complex plane with $z = x + \iota y$ , it must
satisfy the Laplace's equation:
\begin{equation}
 \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} = 0
\end{equation}
If we make a coordinate transformation $z' = x' + \iota y' = f(z)$, $\phi'(z')$ also satisfies the Laplace's equation in the new coordinate system:
\begin{equation}
 \frac{\partial^2 \phi'}{\partial x'^2} + \frac{\partial^2 \phi'}{\partial y'^2} = 0
\end{equation}
provided that the transformation f z is analytic everywhere in the region under
consideration. Hence, in electrostatics, a CT preserves the potential in each coordinate system:
\begin{equation}
\phi'(x', y') = \phi(x,y)
\end{equation}
Moreover, the preservation of local angles ensures that the boundary conditions in the transformed space remain unchanged. Thus, the dielectric constant of each material is also conserved:
\begin{equation}
\epsilon'(x',y') = \epsilon(x,y)
\end{equation}
Since both the electrostatic potential and the material permittivity are preserved under 2D conformal mapping, the delicate design of a metamaterial with a spatial variation in its constitutive parameters is no longer necessary. This not only simplifies the fabrication process, but also provides an easy route to engineering the plasmonic properties of the transformed nanostructures.

## 3. Creating a Singular Metasurface

A 3D system (Fig. 1A), periodic in one of the dimensions and translationally invariant in the two other directions. The Fig. 1 blue-shaded areas are metallic and support surface plasmons whose spectrum is characterized by three wave-vectors: $k_x$, $k_y$, $k_z$.

Our intent is to show that the x dimension can be hidden by using 2D conformal transformations in which the x, y coordinates are represented by a complex number $z_1 = x_1 + \iota y_1$. Conformal transformations in two dimensions have the property of conserving the permittivity and permeability, $\epsilon$, $\mu$ in the plane of the transformation so that in this plane, we are working with the same materials in all coordinate frames. Under each successive transformation, the spectral properties are preserved, and the modes
once calculated in the initial frame can be found in the other frames through the properties of the transformation.
In the first step, we compress $x_1 = -\infty$ to a point at the origin $z_2 = a*exp(2\pi z_1/d)$. This transformation
produces knife edges. Next, we compress $z_2 = +\infty $ to a point by inverting
about $z_2 = -a$ that is finite in the $x_3$, $y_3$ plane, $z_3 = \frac{1}{z_2 -a}$. Last, we add a further transformation
\begin{equation}
z_4 = \frac{T}{2\pi} \ln{\left (z_3 + \frac{1}{2a}\right)}
\end{equation}
to create a periodic metasurface.


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

from scipy.interpolate import griddata
from matplotlib import cm
from scipy.optimize import fsolve
from scipy.optimize import minimize
import math
import cmath
import scipy.integrate as integrate
from collections import Counter
```

```python
def conformal_mapping1(z, d):
    return a * np.exp(2 * np.pi * z / d)

def conformal_mapping2(z):
    return np.divide(1, z - a)

def conformal_mapping3(z):
    return (100 / (2 * np.pi))*np.log(( z + 1 / (2 * a)))

a = 1
d1 = 2
d2 = 2
d3 = 2
d = 6

y1_range = np.linspace(-d2 - d3, -d2, 1000)
y2_range = np.linspace(d1, d1 + d3, 1000)

x = np.linspace(-10, 10, 1000)
X1, Y1 = np.meshgrid(x, y1_range)
X2, Y2 = np.meshgrid(x, y2_range)
Z1 = X1 + 1j * Y1
Z2 = X2 + 1j * Y2

first_map_Z1 = conformal_mapping1(Z1, d)
first_map_Z2 = conformal_mapping1(Z2, d)

second_map_Z1 = conformal_mapping2(first_map_Z1)
second_map_Z2 = conformal_mapping2(first_map_Z2)

third_map_Z1 = conformal_mapping3(second_map_Z1)
third_map_Z2 = conformal_mapping3(second_map_Z2)

plt.figure(figsize=(36, 8))

plt.subplot(1, 4, 1)
plt.title("Z1 and Z2")
plt.plot(Z1.real, Z1.imag, 'b.')
plt.plot(Z2.real, Z2.imag, 'r.')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.subplot(1, 4, 2)
plt.title("First Map")
plt.plot(first_map_Z1.real, first_map_Z1.imag, 'b.')
plt.plot(first_map_Z2.real, first_map_Z2.imag, 'r.')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.subplot(1, 4, 3)
plt.title("Second Map")
plt.plot(second_map_Z1.real, second_map_Z1.imag, 'b.')
plt.plot(second_map_Z2.real, second_map_Z2.imag, 'r.')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.subplot(1, 4, 4)
plt.title("Third Map")
plt.plot(third_map_Z1.real, third_map_Z1.imag, 'b.')
plt.plot(third_map_Z2.real, third_map_Z2.imag, 'r.')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.tight_layout()
plt.show()
```

## 4. Source fields in the transformed space


In this study, we focus on the case of a p-polarized (TM polarized) plane wave, where the magnetic field moves out of the plane ($H_z$), incident on the singular metasurface. This choice is made because any more complex wave front can be expressed as a superposition of plane waves. Now, when we transform the geometry of the system, we must also adjust the form of the source.

In the metasurface frame, we create an incident wave using a periodic arrangement of magnetic current line sources ("monopoles") positioned on the right-hand side of the metasurface. These sources, assumed to be located infinitely far away to disregard their immediate surroundings, emit a plane wave directed towards the surface. We make an important **assumption** here: the **period of the metasurface is significantly smaller than the wavelength in free space**.

When the incident wave encounters the singular surface, it results in reflected and transmitted waves. Each of these three wave types contributes to the excitation of surface plasmons at the singular surface. To ensure the proper behavior of these waves, we need additional elements: one sink positioned infinitely far away to collect the reflected waves and another sink placed at minus infinity to capture the transmitted waves.

To determine the source in the slab frame, we utilize two key observations:
1. a magnetic current line remains conserved under the transformation
2. sources located at $+\infty$ in the metasurface frame correspond to specific points in the slab frame ($z_1 = \iota nd$), while those at $-\infty$ map to different points ($z_1 = \iota(n+1/2 )d$, where n is an integer).

Consequently, the monopole sources generating the incident wave and those receiving the reflected waves find their place in the air region of the slab frame, while the monopoles receiving the transmitted wave are positioned within the metal region.

#### Magnetic field of the **incident, reflected and transmitted waves in the metasurface frame**


Magnetic field of the **incident, reflected and transmitted waves in the metasurface frame** is given as,


\begin{equation}
\left\{
{\begin{array}{l}
H_z^{inc} = H_0 e^{-i k_{0x} x_4 + i k_{0y} y_4}\\
H_z^{ref} = r H_0 e^{i k_{0x} x_4 + i k_{0y} y_4}\\
H_z^{tra} = t H_0 e^{-i k_{0x}^{'}  x_4 + i k_{0y} y_4}
\end{array}}
\right.
\end{equation}

where $H_0$ is the wave amplitude, $r$ and $t$ are reflection and transmission coefficients, $k_{0x}= \sqrt{k_0^2-k_{0y}^2}$ and $k_{0x}^{'}= \sqrt{\varepsilon k_0^2-k_{0y}^2}$.

#### Magnetic field of the **incident, reflected and transmitted waves in the slab frame**

Starting from plane waves in the metasurface frame, we apply the transformation equation and write the source fields in the slab geometry. We assume that the metasurface is subwavelength and that the spatial region of interest satisfies $T \ll |x_4| \ll \lambda$. Under this assumption, the transformation reduces to
\begin{equation}
\begin{split}
z_4 & \approx -\frac{T}{2\pi} \mathrm{ln} \bigg( \frac{\pi}{d}z_1 \bigg)
\end{split}
\end{equation}
for the incident and reflected waves on the right-hand side of the singular surface ($T \ll x_4 \ll \lambda$). For transmitted wave on the left side of singular surface we have $T \ll -x_4 \ll \lambda$ and,
\begin{equation}
\begin{split}
z_4 & \approx \frac{T}{2\pi} \mathrm{ln} \bigg( \frac{\pi}{d}(z_1 \pm i \frac{d}{2}) \bigg).
\end{split}
\end{equation}

Then, the incident wave can be written as

\begin{equation}
\begin{split}
H_z^{inc} &= H_0 e^{-i k_{0x} x_4 + i k_{0y} y_4}\\
&= H_0 e^{-i k_{0x} \frac{z_4+z_4^*}{2} + i k_{0y} \frac{z_4-z_4^*}{2 i}}\\
&= H_0 e^{\frac{-i k_{0x} + k_{0y}}{2} z_4 + \frac{-i k_{0x} - k_{0y}}{2} z_4^*}\\
&\approx H_0 e^{- \frac{-i k_{0x} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1 \big)} \times e^{- \frac{-i k_{0x} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1^* \big)}\\
&\approx H_0 \bigg(1 + \frac{i k_{0x} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1 \big) \bigg) \times \bigg(1 + \frac{i k_{0x} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1^* \big) \bigg) \\
&\approx H_0 \bigg(1 + \frac{i k_{0x} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1 \big) + \frac{i k_{0x} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}z_1^* \big) \bigg) \\
&= H_0 \bigg(1 + i\frac{k_{0x}T}{4\pi} \mathrm{ln} \big( (\frac{\pi}{d})^2 |z_1|^2 \big) + \frac{k_{0y}T}{4\pi} \mathrm{ln} \big( \frac{z_1^*}{z_1} \big) \bigg) \\
&= H_0 \bigg(1 + i\frac{k_{0x}T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) + i\frac{k_{0x}T}{4\pi} \mathrm{ln} \big(|z_1|^2 \big) + \frac{k_{0y}T}{4\pi} \mathrm{ln} \big( \frac{z_1^*}{z_1} \big) \bigg)\\
&= H_0 \bigg(1 + i\frac{k_{0x}T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) - i\frac{k_{0x}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x + \frac{k_{0y}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{\text{sgn}(y_1)k_x} e^{i k_x x_1} d k_x \bigg)\\
&= H_0 \bigg(1+i \frac{k_{0x} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) +  \int\limits_{-\infty}^{\infty} a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x +  \int\limits_{-\infty}^{\infty} a_s \frac{e^{-|k_x||y_1|}}{\text{sgn}(y_1)k_x} e^{i k_x x_1} d k_x,
\end{split}   
\end{equation}

where we have used
\begin{equation}
\begin{split}
\mathrm{ln} \big( |z_1|^2 \big) = -\int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x
\end{split}   
\end{equation}
and
\begin{equation}
\begin{split}
\mathrm{ln} \big( \frac{z_1^*}{z_1} \big) = \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{\text{sgn}(y_1)k_x} e^{i k_x x_1} d k_x.
\end{split}   
\end{equation}

The reflected wave can be easily obtained by replacing $k_{0x}$ with $-k_{0x}$,
\begin{equation}
\begin{split}
H_z^{ref} &= r H_0 e^{i k_{0x} x_4 + i k_{0y} y_4}\\
&= r H_0 \bigg(1 - i\frac{k_{0x}T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) + i\frac{k_{0x}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x + \frac{k_{0y}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1|}}{\text{sgn}(y_1)k_x} e^{i k_x x_1} d k_x \bigg)\\
&= r H_0 \bigg(1-i \frac{k_{0x} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) -  \int\limits_{-\infty}^{\infty} r a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x +  \int\limits_{-\infty}^{\infty} r a_s \frac{e^{-|k_x||y_1|}}{\text{sgn}(y_1)k_x} e^{i k_x x_1} d k_x
\end{split}   
\end{equation}

For the transmitted wave, $k_{0x}^{'} = \sqrt{\varepsilon k_0^2 - k_{0y}^2}$, and we have
\begin{equation}
\begin{split}
H_z^{tra} &= t H_0 e^{-i k_{0x}^{'}  x_4 + i k_{0y} y_4}\\
&= t H_0 e^{-i k_{0x}^{'} \frac{z_4+z_4^*}{2} + i k_{0y} \frac{z_4-z_4^*}{2 i}}\\
&= t H_0 e^{\frac{-i k_{0x}^{'} + k_{0y}}{2} z_4 + \frac{-i k_{0x}^{'} - k_{0y}}{2} z_4^*}\\
&\approx t H_0 e^{\frac{-i k_{0x}^{'} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1 + i \frac{d}{2}) \big)} \times e^{ \frac{-i k_{0x}^{'} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1 + i \frac{d}{2})^* \big)}\\
&\approx t H_0 \bigg(1 - \frac{i k_{0x}^{'} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1+i\frac{d}{2}) \big) \bigg) \times \bigg(1 - \frac{i k_{0x}^{'} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1+i\frac{d}{2})^* \big) \bigg) \\
&\approx t H_0 \bigg(1 - \frac{i k_{0x}^{'} - k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1+i\frac{d}{2}) \big) - \frac{i k_{0x}^{'} + k_{0y}}{2} \frac{T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d}(z_1+i\frac{d}{2})^* \big) \bigg) \\
&= t H_0 \bigg(1 - i\frac{k_{0x}^{'}T}{4\pi} \mathrm{ln} \big( (\frac{\pi}{d})^2 |z_1+i\frac{d}{2}|^2 \big) - \frac{k_{0y}T}{4\pi} \mathrm{ln} \big( \frac{(z_1+i\frac{d}{2})^*}{(z_1+i\frac{d}{2})} \big) \bigg) \\
&= t H_0 \bigg(1 - i\frac{k_{0x}^{'}T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) - i\frac{k_{0x}^{'}T}{4\pi} \mathrm{ln} \big( |z_1+i\frac{d}{2}|^2 \big) - \frac{k_{0y}T}{4\pi} \mathrm{ln} \big( \frac{(z_1+i\frac{d}{2})^*}{(z_1+i\frac{d}{2})} \big) \bigg) \\
&= t H_0 \bigg(1 - i\frac{k_{0x}^{'}T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) + i\frac{k_{0x}^{'}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1+\frac{d}{2}|}}{|k_x|} e^{i k_x x_1} d k_x - \frac{k_{0y}T}{4\pi} \int\limits_{-\infty}^{\infty} \frac{e^{-|k_x||y_1+\frac{d}{2}|}}{\text{sgn}(y_1+\frac{d}{2})k_x} e^{i k_x x_1} d k_x \bigg)\\
&= t H_0 \bigg(1 - i \frac{k_{0x}^{'} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) - \int\limits_{-\infty}^{\infty} t \frac{k_{0x}^{'}}{k_{0x}}  a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x  -  \int\limits_{-\infty}^{\infty} t a_s \frac{e^{-|k_x||y_1+\frac{d}{2}|}}{\text{sgn}(y_1+\frac{d}{2})k_x} e^{i k_x x_1} d k_x
\end{split}   
\end{equation}


Hence, the **source field in the slab frame**:

\begin{equation}
\left\{
{\begin{array}{l}
H_z^{inc} = H_0 \bigg(1+i \frac{k_{0x} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) +  \int\limits_{-\infty}^{\infty} a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x +  \int\limits_{-\infty}^{\infty} a_s \frac{e^{-|k_x||y_1|}}{sgn(y_1)k_x} e^{i k_x x_1} d k_x\\
H_z^{ref} = r H_0 \bigg(1-i \frac{k_{0x} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) -  \int\limits_{-\infty}^{\infty} r a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x +  \int\limits_{-\infty}^{\infty} r a_s \frac{e^{-|k_x||y_1|}}{sgn(y_1)k_x} e^{i k_x x_1} d k_x\\
H_z^{tra} = t H_0 \bigg(1 - i \frac{k_{0x}^{'} T}{2\pi} \mathrm{ln} \big( \frac{\pi}{d} \big) \bigg) - \int\limits_{-\infty}^{\infty} t \frac{k_{0x}^{'}}{k_{0x}}  a_a \frac{e^{-|k_x||y_1|}}{|k_x|} e^{i k_x x_1} d k_x  -  \int\limits_{-\infty}^{\infty} t a_s  \frac{e^{-|k_x||y_1+\frac{d}{2}|}}{sgn(y_1+\frac{d}{2})k_x} e^{i k_x x_1} d k_x
\end{array}}
\right.
\end{equation}

where the source-field is decomposed into two parts:
1. anti-symmetric source (corresponds to waves with the electric field parallel to the metasurface), with amplitude
\begin{equation}
a_a = -\frac{\iota k_{0x}T}{4\pi}H_0
\end{equation}
2. symmetric source (electric field normal to the metasurface), with amplitude
\begin{equation}
a_s = \frac{ k_{0y}T}{4\pi}H_0
\end{equation}

The symmetry of the modes is defined by the parity of $E_x(y)$ in the slab frame: the odd (even) function corresponds to the anti-symmetric (symmetric) mode. At normal incidence, only the $a_a$ mode is excited.

Hence, from above **source field in k-space** for the antisymmetric mode is,
\begin{equation}
    H_z^a(k_x)=       
        \left\{
        {\begin{array}{lr}
         (1-r)a_a \frac{e^{-|k_x||y|}}{|k_x|},&{ - {d_2} < y < {d_1}} \\
         -t \frac{k_{0x}^{'}}{k_{0x}}  a_a \frac{e^{-|k_x||y+\frac{d}{2}|}}{|k_x|},&{ - {(d_2+d_3)} < y < {-d_2}}
        \end{array}}
        \right.
\end{equation}
and for the symmetric mode is,
\begin{equation}
       H_z^s(k_x)= \left\{
        {\begin{array}{lr}
         (1+r)a_s \text{sgn}(k_x) \frac{e^{-|k_x||y|}}{\text{sgn}(y)|k_x|},&{ - {d_2} < y < {d_1}} \\
         -t a_s \text{sgn}(k_x) \frac{e^{-|k_x||y+\frac{d}{2}|}}{\text{sgn}(y+\frac{d}{2})|k_x|},&{ - {(d_2+d_3)} < y < {-d_2}}
        \end{array}}
        \right.
\end{equation}
in which the constant field components are ignored since they do not contribute to the excitation of surface plasmon polaritons.

## 5. Surface Plasmons Dispersion Relation

Upon excitation of a SPP mode at the interface between metal and air, the total (excited) field distribution in the k-space within the slab frame can be expressed as:
\begin{equation}
H_z(k_x)=\left\{{\begin{array}{lr}
(1-r)a_a \frac{e^{-|k_x||y|}}{|k_x|} + (1+r)a_s \frac{e^{-|k_x||y|}}{\text{sgn}(y)k_x}+ {b_ +}e^{-|k_x|y} + {b_ -}e^{|k_x|y},&{ -d_2 < y < d_1}\\
-t \frac{k_{0x}^{'}}{k_{0x}}  a_a \frac{e^{-|k_x||y+\frac{d}{2}|}}{|k_x|} -t a_s \frac{e^{-|k_x||y+\frac{d}{2}|}}{\text{sgn}(y+\frac{d}{2})k_x} + {c_ +}e^{-|k_x|y} + {c_ -}e^{|k_x|y},&{ - ({d_2} + {d_3}) < y <  - {d_2}}
\end{array}} \right.
\end{equation}

where $b_+$, $b_-$ are the scattering coefficients in the dielectric and $c_+$, $c_-$ are the scattering coefficients inside the metal, in k-space.

In order to calculate the excited mode coefficients ($b_+$, $b_-$, $c_+$, $c_-$) of the excited SPPs, we apply the boundary conditions in the slab frame: matching the tangential field components, $H_z$ and $E_x$, at the boundaries $y=d_1$, $y=-d_2$, $y=-(d_2+d_3)$.

In the matrix form, the equations system reads as,
\begin{equation}
\begin{split}
    &\left( {\begin{array}{*{20}{l}}
    e^{|k_x|d_2} & e^{-|k_x|d_2} & -e^{|k_x|d_2} & -e^{-|k_x|d_2} \\
    e^{-|k_x|d_1} & e^{|k_x|d_1} & -e^{|k_x|(d_2+d_3)} & -e^{-|k_x|(d_2+d_3)} \\
    |k_x|e^{|k_x|d_2} & -|k_x|e^{-|k_x|d_2} & -\frac{|k_x|e^{|k_x|d_2}}{\varepsilon} & \frac{|k_x|e^{-|k_x|d_2}}{\varepsilon} \\
    |k_x|e^{-|k_x|d_1} & -|k_x|e^{|k_x|d_1} & -\frac{|k_x|e^{|k_x|(d_2+d_3)}}{\varepsilon} & \frac{|k_x|e^{-|k_x|(d_2+d_3)}}{\varepsilon}
    \end{array}} \right)
    \left( {\begin{array}{*{20}{l}}
    b_+ \\ b_- \\ c_+ \\ c_-
    \end{array}} \right)
    \\
    &=
    \left( {\begin{array}{*{10}{c}}
    -\frac{e^{-|k_x|d_2}}{|k_x|}\\ -\frac{e^{-|k_x|d_1}}{|k_x|} \\ e^{-|k_x|d_2} \\ - e^{-|k_x|d_1}
    \end{array}} \right) a_a (1-r)
    +
    \left( {\begin{array}{*{10}{c}}
    -\frac{e^{-|k_x|(-d_2+\frac{d}{2})}}{|k_x|}\\ -\frac{e^{-|k_x|(d_2+d_3-\frac{d}{2})}}{|k_x|} \\ -\frac{e^{-|k_x|(-d_2+\frac{d}{2})}}{\varepsilon} \\ \frac{e^{-|k_x|(d_2+d_3-\frac{d}{2})}}{\varepsilon}
    \end{array}} \right) a_a \frac{k_{0x}^{'}}{k_{0x}} t
    \\
    &+
    \left( {\begin{array}{*{10}{c}}
    \frac{e^{-|k_x|d_2}}{k_x}\\ -\frac{e^{-|k_x|d_1}}{k_x} \\ - \text{sgn}(k_x) e^{-|k_x|d_2} \\ - \text{sgn}(k_x) e^{-|k_x|d_1}
    \end{array}} \right) a_s (1+r)
    +
    \left( {\begin{array}{*{10}{c}}
    -\frac{e^{-|k_x|(-d_2+\frac{d}{2})}}{k_x}\\ \frac{e^{-|k_x|(d_2+d_3-\frac{d}{2})}}{k_x} \\
    - \text{sgn}(k_x) \frac{e^{-|k_x|(-d_2+\frac{d}{2})}}{\varepsilon} \\
    - \text{sgn}(k_x) \frac{e^{-|k_x|(d_2+d_3-\frac{d}{2})}}{\varepsilon}
    \end{array}} \right) a_s t
\end{split}
\end{equation}
where the Bloch wave condition has been used because of the periodicity of the slab array. Since the system of equations is linear, the response functions $b$ and $c$ can be decomposed into anti-symmetric and symmetric components. For the anti-symmetric excitation ($a_a \neq 0$, $a_s = 0$), we have
\begin{equation}
\begin{split}
        b_{a+} =& \frac{(\varepsilon -1)e^{|k_x| d_3}+\varepsilon +1}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a (1-r) \\
        & -\frac{2 e^{\frac{1}{2}|k_x| d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a \frac{k_{0x}^{'}}{k_{0x}} t \\
        b_{a-} =& \frac{(\varepsilon -1)e^{|k_x| d_3}+\varepsilon +1}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a (1-r) \\
        & -\frac{2 e^{\frac{1}{2}|k_x| d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a \frac{k_{0x}^{'}}{k_{0x}} t
\end{split}
\end{equation}
\begin{equation}
\begin{split}
        c_{a+} =& \frac{2\varepsilon}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a (1-r) \\
        & + \frac{e^{-\frac{1}{2}|k_x| d}((\varepsilon-1)e^{2|k_x| d_2}-(\varepsilon+1))}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a \frac{k_{0x}^{'}}{k_{0x}} t \\
        c_{a-} =& \frac{2 \varepsilon  e^{|k_x| d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a (1-r) \\
        & + \frac{e^{\frac{1}{2}|k_x| d}((\varepsilon-1)e^{2|k_x| d_2}-(\varepsilon+1))}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) + (\varepsilon+1)(e^{|k_x|d}-1))} a_a \frac{k_{0x}^{'}}{k_{0x}} t
\end{split}
\end{equation}

On the other hand, the symmetric excitation ($a_a = 0$, $a_s \neq 0$) gives
\begin{equation}
\begin{split}
        b_{s+} =& \frac{(\varepsilon-1)e^{|k_x|d_3} - (\varepsilon+1)}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s (1+r) \\
        & + \frac{2 e^{\frac{1}{2}|k_x|d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))}\text{sgn}(k_x) a_s t \\
        b_{s-} =& -\frac{(\varepsilon-1)e^{|k_x|d_3} - (\varepsilon+1)}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s (1+r) \\
        & -\frac{2 e^{\frac{1}{2}|k_x|d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))}\text{sgn}(k_x) a_s t
\end{split}
\end{equation}
\begin{equation}
\begin{split}  
        c_{s+} =& -\frac{2\varepsilon}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s (1+r) \\
        & +\frac{e^{-\frac{1}{2}|k_x|d}((\varepsilon-1)e^{2 d_2 |k_x|}+ (\varepsilon+1))}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s t \\
        c_{s-} =& \frac{2\varepsilon e^{|k_x|d}}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s (1+r) \\
        & -\frac{e^{\frac{1}{2}|k_x|d}((\varepsilon-1)e^{2 d_2 |k_x|}+ (\varepsilon+1))}{|k_x| ((\varepsilon-1)(e^{|k_x|(d_1+d_2)} - e^{|k_x|d_3}) - (\varepsilon+1)(e^{|k_x|d}-1))} \text{sgn}(k_x) a_s  t
\end{split}
\end{equation}



Dispersion relation of surface plasmon excitation can be obtained by looking at the poles of the scattering coefficients, that is, through the condition that $b_+$ and $b_-$ diverge. Focusing our attention on the plasmon pole, we disregard the pole at $k_x = 0$ due to its nature as a branch point, representing a localized virtual excitation rather than SPPs.
1. For anti-symmetric (odd) mode  
\begin{equation}
\begin{split}
(\varepsilon-1)(e^{|k_{x}|(d_1+d_2)} - e^{|k_{x}|d_3}) + (\varepsilon+1)(e^{|k_{x}|d}-1)=0
\end{split}
\end{equation}
2. for the symmetric (even) mode
\begin{equation}
\begin{split}
(\varepsilon-1)(e^{|k_{x}|(d_1+d_2)} - e^{|k_{x}|d_3}) - (\varepsilon+1)(e^{|k_{x}|d}-1)=0
\end{split}
\end{equation}

For the wedge (groove) metasurface, the symmetric (antisymmetric) band spans a frequency range $\omega_{c1}\le\omega<\omega_{sp}$ while the antisymmetric (symmetric band) spans $\omega_{sp}<\omega\le \omega_{c2}$.  Here, $\omega_{sp}$ is the surface plasmon frequency, and $\omega_{c(1,2)}$ are the lower and upper cut-off frequencies.

The dispersion relation for both the symmetric and antisymmetric modes is plotted below. For the case of normal incidence, only antisymmetric mode is excited.

```python
x = np.linspace(0.1,50,100)
A = np.exp(0.9*x) - np.exp(0.1*x)
B = np.exp(x) - 1
epsilon_as = (A-B)/(A+B)
epsilon_s = (A+B)/(A-B)

y_as = np.sqrt(1/(1-epsilon_as))
y_s = np.sqrt(1/(1-epsilon_s))
```

In the most general case, both the symmetric and antisymmetric bands can be excited  at an oblique incidence. However, at normal incidence only the antisymmetric band is excited. It should be noted that the bands have lower and upper cut-off frequencies different from 0 and $\omega_p$. For the wedge (groove) metasurface, the symmetric (antisymmetric) band spans a frequency range $\omega_{c1} \leq \omega < \omega_{sp}$ while the antisymmetric (symmetric band) spans  $\omega_{sp} \leq \omega < \omega_{c2}$ Here, $\omega_{sp}$ is the surface plasmon frequency, and $\omega_{c1,2}$ are the lower and upper cut-off frequencies. The relations for these cutoff frequencies have been derived in the previous works.

Also note that the above figures have been generated for lossless media (which means $\epsilon$ is real). For the rest of the report, we consider a finite damping.

Once the dispersion relation has been obtianed, we now obtain the plasmon poles using this relation. $k_{px}$ is obtained as a solution to the complex equations
    \begin{equation}
        (\epsilon - 1)(e^{k_{px} (d_1 + d_2)} - e^{k_{px}d_3}) + (\epsilon + 1)(e^{k_{px}d } -1) = 0
    \end{equation}
   for the antisymmetric case and
   \begin{equation}
        (\epsilon - 1)(e^{k_{px} (d_1 + d_2)} + e^{k_{px}d_3}) + (\epsilon + 1)(e^{k_{px}d } -1) = 0
    \end{equation}
    for the symmetric case.
    
We have,
        $\epsilon = 1-\dfrac{\omega_p^2}{\omega(\omega + \iota \Gamma)}$ where $\omega_p = 8.95eV/\hbar$ and $\Gamma = 65.8meV/\hbar$.

We use Newton-Raphson method to solve this non-linear equation.

```python

def f(k, wp, w, y):
    return -wp**2 / (w * (w + 1j * y)) * (np.exp(0.1*np.sqrt(k**2)) - np.exp(0.9 * np.sqrt(k**2))) + (2 - wp**2 / (w * (w + 1j * y))) * (np.exp(1 * np.sqrt(k**2)) - 1)

def df(k, wp, w, y):
    return -wp**2 / (w * (w + 1j * y)) * (np.exp(0.1*np.sqrt(k**2)) - 0.9 * np.exp(0.9 * np.sqrt(k**2))) + 1 * (2 - wp**2 / (w * (w + 1j * y))) * np.exp(1 * np.sqrt(k**2)) * np.sign(k)

rvalues = np.linspace(-5, 5, 100)
ivalues = np.linspace(-5, 5, 100)

wp = 8.95
w = 0.6 * wp
y = 65.8 * 10**(-3)

delta = 1e-6

roots = []

for r in rvalues:
    for i in ivalues:
        guess = r + i * 1j
        for n in range(1000):
            next_guess = guess - f(guess, wp, w, y) / df(guess, wp, w, y)
            if abs(next_guess - guess) < delta:
                already_in = False
                for root in roots:
                    if abs(next_guess - root) < delta: 
                        already_in = True 
                        break
                if not already_in:
                    roots.append(next_guess)
                break
            guess = next_guess

# Print the roots
```

```python
print("Roots:", [i/10 for i in roots])
```

The roots for asymmetric case ($\omega = 0.6 \omega_p$) been printed above. The error tolerance is taken to be $10^{-10}$. We notice that the roots are close enough so that we can truncate their values upto 3 decimal places for later use. For a better visualization, we plot these roots below.

```python
plt.figure(figsize = (10,6))
plt.scatter(np.real([i/10 for i in roots]), np.imag([i/10 for i in roots]), color='red', label='Roots')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Anti-Symmetric')
plt.grid(True)
plt.legend()

plt.ylim(-5,5)
plt.xlim(-1.5,2)
plt.show()
```

For further sections, we use $k_{px} = -1.27 + \iota 3.099$ (discarding the solution with positive imaginary part so that the fields vanish at infinity) for the asymmetric mode. A similar calculation for the symmetric mode at $\omega = 0.8\omega_p$ gives $-1.272 + \iota 9.456$. 

## 6. **Local Calculation**
Without spatial dispersion in the metal, the field can be represented merely by a divergence-free magnetic field
\begin{equation}
H_z(k_x)=\left\{{\begin{array}{lr}
(1-r)a_a \frac{e^{-|k_x||y|}}{|k_x|} + (1+r)a_s \frac{e^{-|k_x||y|}}{\text{sgn}(y)k_x}+ {b_ +}e^{-|k_x|y} + {b_ -}e^{|k_x|y},&{ -d_2 < y < d_1}\\
-t \frac{k_{0x}^{'}}{k_{0x}}  a_a \frac{e^{-|k_x||y+\frac{d}{2}|}}{|k_x|} -t a_s \frac{e^{-|k_x||y+\frac{d}{2}|}}{\text{sgn}(y+\frac{d}{2})k_x} + {c_ +}e^{-|k_x|y} + {c_ -}e^{|k_x|y},&{ - ({d_2} + {d_3}) < y <  - {d_2}}
\end{array}} \right.
\end{equation}

In calculation of the excited SPP mode in real space, we have two approaches: the discrete k-method and the continuous-k method.

For the continuous-k method, we calculate the field generated by a single monopole and then sum the field of all monopoles together. In the discrete-k method, the periodicity of the monopole array only excites SPPs with a specific value of the k-vector (k=ng where n is an integer and $g = \frac{2\pi}{d}$ is the reciprocal lattice constant).



The two methods (discrete-k and continuous-k) both give the spectrum of surface conductivity. They are identical when the k-vector of SPP is large, but differ when the k-vector is small (near 0 and $\omega_p$ frequencies). This difference comes from neglecting the branch cut when applying the residue theorem in the continuous-k method. Therefore, the discrete-k method is more accurate for the local calculation.

### (a.) Continuous-k Method

First we consider two monopoles on the x-axis in the slab frame, one for incident and reflected waves and the other for transmitted wave.

From $H_z^0(y, k_x)$, the field $H_z^0(x,y)$ generated by the two monopoles in real space can be obtained by inverse Fourier Transform.

\begin{equation}
H_z^0(x,y) = \int_{-\infty}^\infty H_z^0(k_x, y) e^{i{k_x}x} d{k_x}
\end{equation}

And then, adding together the fields of all monopole pairs with x=0 at $y = nd$, we have $H_z(x, y) = \sum_n H_z^0(x, y-nd)$, and the corresponding electric field can be calculated from the Maxwell's equations as
\begin{equation}
    E_x(x,y) = \frac{i}{\omega \varepsilon_0 \varepsilon} \frac{\partial H_z}{\partial y}
\end{equation} and
\begin{equation}
E_y(x,y) = -\frac{i}{\omega \varepsilon_0 \varepsilon} \frac{\partial H_z}{\partial x}
\end{equation}

Therefore, the total power absorbed is, (trick is to evaluate the power flow at the excitation point)
\begin{equation}
    P_{abs} = 2 \int\limits_{-(d_2+d_3)}^{d_1} S_x dy
\end{equation} where $S_x = \frac{1}{2} \mathrm{Re}[E_y^{*} H_z]$



### (b.) Discrete-k Method

If the field of one monopole is expressed as $H_z^0(x,y)$, the total field can be written as
\begin{equation}
\begin{split}
H_z(x,y) &= \sum_n H_z^0(x,y-n d) \\
&= H_z^0(x,y) * \sum_n \delta(y - n d)
\end{split}
\end{equation}
where $n$ is an integer, $d$ is the period of the monopole source array and $*$ stands for convolution. Since $H_z(x,y)$ is a periodic function, it can be written as a Fourier series
\begin{equation}
\begin{split}
H_z(x,y) &= \sum_n c_n e^{i n g y}
\end{split}
\end{equation}
where
\begin{equation}
\begin{split}
c_n &= \frac{1}{d}\int_{-(d_2+d_3)}^{d_1} H_z(x,y) e^{-i n g y} dy \\
&= \frac{1}{d} \int_{-(d_2+d3)}^{d_1} \sum_n H_z^0(x,y-n d) e^{-i n g y} dy\\
\end{split}
\end{equation}
Therefore,
\begin{equation}
\begin{split}
H_z(x,y) &= \sum_n g H_z^0(x,n g) e^{i n g y} \\
&= \int_{-\infty}^{\infty} g H_z^0(x,k_y) \sum_n \delta(k_y - n g) e^{i k_y y} d k_y
\end{split}
\end{equation}
Hence the field representation in $k$-space is $H_z^0(x,k_y) \times \sum_n g \delta(k_y - n g)$.

Solving the above equation system, we can calculate the field distribution of the excited mode in real space by inverse Fourier transform.
\begin{equation}
    H_z(x,y) = \int\limits_{-\infty}^\infty H_z(k_x,y) e^{i{k_x}x} d{k_x}
\end{equation}

Then the electric field can be calculated as
\begin{equation}
    E_x(x,y) = \frac{i}{\omega \varepsilon_0 \varepsilon} \frac{\partial H_z}{\partial y}
\end{equation} and
\begin{equation}
E_y(x,y) = -\frac{i}{\omega \varepsilon_0 \varepsilon} \frac{\partial H_z}{\partial x}
\end{equation}

Power absorption in one period:
\begin{equation}
    P_{abs} = \int\limits_{slab} \frac{1}{2} \omega \varepsilon_0 \mathrm{Im}[\varepsilon] |E|^2 dx dy = \int\limits_{slab} \frac{1}{2} \omega \varepsilon_0 \mathrm{Im}[\varepsilon] (|E_x|^2 + |E_y|^2) dx dy
\end{equation}

## 6.2 Energy Dissipation by SPPs


Energy is conserved under transformations, hence power absorbed by the concave or convex metasurfaces in the transformed coordinate system $z_4$ is equivalent to the power dissipated in the original geometry $z_1$, dissipation is due to loss in the metal. Power absorption in the slab frame in one period (trick is to evaluate the power flow at the excitation point, x=0):
\begin{equation}
\begin{split}
    P^{(a,s)}_{abs} &= \int\limits_{slab} \frac{1}{2} \omega \varepsilon_0 \mathrm{Im}[\varepsilon] |E|^2 dxdy = \int\limits_{slab} \frac{1}{2} \omega \varepsilon_0 \mathrm{Im}[\varepsilon] (|E_x|^2+|E_y|^2) dxdy\\
    &= \frac{4\pi^2|a|^2|k_{px}|^2 \mathrm{Im}[\varepsilon]}{\omega \varepsilon_0 |\varepsilon|^2} (\frac{|\Lambda_{(a,s)+}|^2}{-2\mathrm{Re}[\sqrt{k_{px}^2}]} (e^{2 \mathrm{Re}[\sqrt{k_{px}^2}]d_2}-e^{2 \mathrm{Re}[\sqrt{k_{px}^2}](d_2+d_3)}) \\
    &+ \frac{|\Lambda_{(a,s)-}|^2}{2\mathrm{Re}[\sqrt{k_{px}^2}]}(e^{-2 \mathrm{Re}[\sqrt{k_{px}^2}]d_2}-e^{-2 \mathrm{Re}[\sqrt{k_{px}^2}](d_2+d_3)}) \big) \frac{1}{\mathrm{Im}[k_{px}]}
\end{split}
\end{equation}
where the coefficients $\Lambda_{(a,s)\pm}$ stand for the antisymmetric and symmetric mode coefficients.

The absorption cross section of the structure can be obtained by normalizing to the input energy on the system in one period,  
\begin{equation}
    \sigma^{(a,s)}_{abs} = \frac{P^{(a,b)}_{abs}}{\frac{1}{2} \sqrt{\frac{\mu_0}{\varepsilon_0}} H_0^2 T \cos{\theta_{in}}},
\end{equation}

## 6.3 Metasurfaces with blunt singularities

Fabricated metasurfaces inevitably deviate from perfect singularity, exhibiting what is termed a "blunt" singularity. In the case of a metasurface with rounded singularities, it effectively transforms into a truncated slab array (or truncated cavity array for the groove geometry). This transition carries a significant implication: when the slabs or cavities are not infinitely extended, the SPP modes excited within them encounter reflection at the terminals (located at x = -L/2 and x = L/2). This reflection leads to the quantization of the supported SPP modes, thereby transforming the continuous spectrum characteristic of a singular metasurface into a discrete spectrum in the presence of blunt singularities.

To assess the field distribution and energy dissipation within the truncated slab or cavity arrays, we must initially compute the reflection coefficient of SPPs at the terminals of these structures. For brevity, we will primarily discuss the cavity array, noting that the same principles apply to truncated slab arrays. The radius of the rounded singularity within the metasurface frame dictates both the length L of the cavities in the slab frame, and the precise configuration of the terminal.

To compute the reflection coefficient of SPP modes at the cavity end, we consider the terminal as a flat vertical interface between air and metal. Subsequently, we analyze the SPP mode field $H_z^{sp}$ within the cavity ($-L/2 < x_1 < L/2$) and the field $H_z^{out}$ outside the cavity region ($x_1 < -L/2$ and $x_1 > L/2$). Imposing the continuity condition of tangential fields and power flow at $x_1 = L/2$, we get
\begin{eqnarray}
(1+r_{sp})H_z^{sp} &=& H_z^{out} \\
(1-r_{sp})E_y^{sp} &=& E_y^{out} \\
\int\limits_{-(d_2+d_3)}^{d_1}(1-r_{sp}^*)(1+r_{sp})E_y^{sp*} H_z^{sp} dy &=& \int\limits_{-(d_2+d_3)}^{d_1} E_y^{out*} H_z^{out} dy
\end{eqnarray}

where, $H_z^{sp}$ and $E_y^{sp}$ are given in Sec. 6.1.

In the region outside the periodic cavity array we expand the magnetic field as a series of Bloch waves,
\begin{equation}
\begin{split}
H_z^{out} = \sum_{g} h(g) e^{i g y}
\end{split}
\end{equation}
where $g=n \frac{2\pi}{d}$ and $n$ is an integer. The coefficients $h(g)$ can be expressed as
\begin{equation}
\begin{split}
h(g) &= \frac{1}{d} \int\limits_{-(d_2+d_3)}^{d_1} H_z^{out} e^{-i g y} dy
= \frac{1+r_{sp}}{d} \int\limits_{-(d_2+d_3)}^{d_1} H_z^{sp} e^{-i g y} dy
= \frac{1+r_{sp}}{d} I(g)
\end{split}
\end{equation}
On the other hand, we can derive the electric field in this region as $E_y^{out} = - \frac{i}{\omega \varepsilon \varepsilon_0} \frac{\partial H_z^{out}}{\partial x}$. This yields
\begin{equation}
\begin{split}
E_y^{out} =  \frac{1}{\omega \varepsilon \varepsilon_0} \sum_{g} \sqrt{\varepsilon k_0^2 - g^2} h(g) e^{i g y}
\end{split}
\end{equation}
where we have used $h(g) \propto e^{i \sqrt{\varepsilon k_0^2 - g^2}x}$ and $\varepsilon$ should be replaced by $1$ for the wedge system. Replacing the expression for $h(g)$, in the above equation we obtain
\begin{equation}
\begin{split}
E_y^{out} = \frac{1+r_{sp}}{\omega \varepsilon \varepsilon_0 d} \sum_{g} \sqrt{\varepsilon k_0^2 - g^2} I(g) e^{i g y}
\end{split}
\end{equation}
Substituting the expressions for $H_z^{out}$ and $E_y^{out}$ into the continuity condition of power flow and after some algebra we arrive to
\begin{equation}
\begin{split}
\frac{1-r_{sp}}{1+r_{sp}} = \frac{1}{ \omega \varepsilon \varepsilon_0 d} \frac{\sum_{g}  \left( \sqrt{\varepsilon k_0^2 - g^2} \right) |I(g)|^2}{\int\limits_{-(d_2+d_3)}^{d_1} E_y^{sp} H_z^{sp^*} dy} \equiv G
\end{split}
\end{equation}
where we take $k_0 = 0$ in the quasi-static limit. From above equation, we obtain the reflection coefficient as
\begin{equation}
    r_{sp} = \frac{1-G}{1+G}
\end{equation}


Since in the quasistatic limit radiative loss is small, the amplitude of the reflection coefficient is $|r_{sp}|\approx 1$. Hence, we consider only the reflection phase, $r_{sp} = e^{i \phi}$, and write the field of the SPP mode in the cavity of length $L$,
\begin{equation}
    H_z(x,y)= \left\{
        {\begin{array}{lr}
        \begin{split}
            & i 2\pi a ({\Gamma_+}e^{-|k_{px}|y}+ {\Gamma_-}e^{|k_{px}|y}) \\
            & \times (e^{i k_{px} |x|} + e^{-i k_{px} |x|+ i k_{px} L + i\phi}) \frac{1}{1 \mp e^{i k_{px} L + i \phi}}
        \end{split}
        ,&{ - {d_2} < y < {d_1}}\\
        \begin{split}
            & i 2\pi a ({\Lambda_+}e^{-|k_{px}|y}+ {\Lambda_-}e^{|k_{px}|y}) \\
            & \times (e^{i k_{px} |x|} + e^{-i k_{px} |x|+ i k_{px} L + i\phi}) \frac{1}{1 \mp e^{i k_{px} L + i \phi}}
        \end{split}
        ,&{ - {d_2+d_3} < y < -{d_2}}
        \end{array}}
        \right.
\end{equation}
in which $\mp$ stands for $-$ for the anti-symmetric source, and $+$ for the symmetric source.

## 7. Modeling a Singular Metasurface with an effective surface conductivity

When no SPP mode is excited on the metasurface, incident waves effectively bypass it, rendering the singular surface akin to a flat surface. However, upon the excitation of SPPs, energy dissipation occurs at the metasurface. This dissipation is characterized by an effective surface conductivity model. Notably, due to the distinct charge distributions at the metasurface, the antisymmetric SPP mode elicits an electric surface current, whereas the symmetric SPP mode prompts a magnetic surface current. Consequently, we model the antisymmetric mode using an effective electric conductivity, $\sigma_e=\sigma_{er} + i \sigma_{ei}$, and the symmetric mode with an effective magnetic conductivity, $\sigma_m=\sigma_{mr} + i \sigma_{mi}$.

For the **anti-symmetric mode**, the effective electric surface current gen-
erated by the excited SPP modes yields the discontinuity of the tangential magnetic field, while the tangential electric field is continuous,
\begin{eqnarray}
    E_y^{inc}+E_y^{ref}-E_y^{t} &=& 0     \\
    H_z^{inc}+H_z^{ref}-H_z^{t} &=& -\sigma_e E_y^{loc}
\end{eqnarray}
Here, $E_y^{loc}$ is the local tangent electric field in the homogenized system, $E_y^{loc} = \frac{1}{2}(E_y^{inc}+E_y^{ref}+E_y^{tra})$. By solving the first equation, we obtain $t = \frac{k_{0x} \varepsilon}{k_{0x}^{'}}(1-r)$, so we write
\begin{equation}
    \begin{split}
      E_y^{loc} &= -(1-r)\frac{k_{0x}}{\omega \varepsilon_0} H_0  
    \end{split}
\end{equation}
where we see that the amplitude of the anti-symmetric mode is $\propto 1-r$. The primed coefficients $\Gamma_{(x,y)\pm}^{'}$ and $\Lambda_{(x,y)\pm}^{'}$ are related to $\Gamma_{(x,y)\pm}$ and $\Lambda_{(x,y)\pm}$ by
\begin{equation}
\begin{split}
\Gamma_{a\pm} &= (1-r)\Gamma_{a\pm}^{'} \\
\Lambda_{a\pm} &= (1-r)\Lambda_{a\pm}^{'} \\
\end{split}
\end{equation}





Similarly, for the **symmetric mode** we have an effective magnetic surface current which yields,
\begin{eqnarray}
    E_y^{inc}+E_y^{ref}-E_y^{t} &=& -\sigma_m H_z^{loc}     \\
    H_z^{inc}+H_z^{ref}-H_z^{t} &=& 0
\end{eqnarray}
where the local tangent magnetic field in the homogenized system is $H_z^{loc} = \frac{1}{2}(H_z^{inc}+H_z^{ref}+H_z^{tra})$. In this case we have $t=1+r$, so
\begin{equation}
    \begin{split}
      H_z^{loc} &= (1+r) H_0  
    \end{split}
\end{equation}
and the amplitude of the symmetric mode is $\propto 1+r$. Normalized mode coefficients can be defined as
\begin{equation}
\begin{split}
\Gamma_{s\pm} &= (1+r)\Gamma_{s\pm}^{'} \\
\Lambda_{s\pm} &= (1+r)\Lambda_{s\pm}^{'} \\
\end{split}
\end{equation}


These derivations highlight that the antisymmetric mode amplitude is proportional to the local electric field $E_y^{loc}$, while the symmetric mode amplitude is proportional to the local magnetic field $H_z^{loc}$. This rationale justifies the introduction of two types of surface conductivities, electric and magnetic, in our model to mimic energy dissipation by the excited SPP wave. With this approach, the complex singular metasurface is simplified into an easier boundary value problem.

## 8. Reflection Spectrum of the Singular Metasurface

Consider a symmetric metasurface and a normally incident plane wave as a source such that only the antisymmetric band is excited. Correspondingly, in this
 scenario themetasurface is modeled by only an effective
 electric conductivity.The calculated conductivity is presented
 in the figure.

```python
y = 65.8 * 10**(-3)
wp = 8.95
k0 = 0.2
T = 1 # 10nm
z0 = 377 # 377 ohm
```

```python
d = 1.0 # DEFINE
d3 = 0.9*d
d1 = 0.5*(d-d3)
d2 = 0.5*(d-d3)
```

```python
# Finding the suitable value of kx for a given w

def f(k, wp, w, y):
    return -wp**2 / (w * (w + 1j * y)) * (np.exp(np.sqrt(k**2)*0.1) - np.exp(0.9 * np.sqrt(k**2))) + (2 - wp**2 / (w * (w + 1j * y))) * (np.exp(1 * np.sqrt(k**2)) - 1)

def df(k, wp, w, y):
    return -wp**2 / (w * (w + 1j * y)) * 0.1*(np.exp(np.sqrt(k**2)*0.1) - 0.9 * np.exp(0.9 * np.sqrt(k**2))) + 1 * (2 - wp**2 / (w * (w + 1j * y))) * np.exp(1 * np.sqrt(k**2)) * np.sign(k)

def find_kpx(v):
    rvalues = np.linspace(-5, 5, 20)
    ivalues = np.linspace(-5, 5, 20)

    w = v * wp

    delta = 1e-6

    roots = []

    for r in rvalues:
        for i in ivalues:
            guess = r + i * 1j
            for n in range(1000):
                next_guess = guess - f(guess, wp, w, y) / df(guess, wp, w, y)
                if abs(next_guess - guess) < delta:
                    already_in = False
                    for root in roots:
                        if abs(next_guess - root) < delta: 
                            already_in = True 
                            break
                    if not already_in and abs(next_guess) != 0 and abs(next_guess.real) !=0:
                        roots.append(next_guess)
                    break
                guess = next_guess
    
    k_r = [-abs(round(c.real, 4)) for c in roots]
    k_i = [-abs(round(c.imag, 4)) for c in roots]

    # Create a new rounded complex number
    k = [complex(k_r[i], k_i[i]) for i in range(len(roots))]
    
    counts = Counter(k)

    k_root, count = counts.most_common(1)[0]
    return k_root
```

```python
# Functions for real and imaginary parts of permittivity, and excited field coefficients

def eps_r(v):
    return 1 - 1/(v**2 + y**2/wp**2)

def eps_i(v):
    return (y/(v*wp))/(v**2 + y**2/wp**2)

def lam_dash_a_pos(v, k_px):
    e = complex(eps_r(v), eps_i(v))
    kpx = cmath.sqrt(k_px**2)
    return (2*e)/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))  +  (e*cmath.exp(-0.5 * kpx*d)*((e-1)*cmath.exp(2*d2*kpx) - (e+1)))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))

def lam_dash_a_neg(v, k_px):
    e = complex(eps_r(v), eps_i(v))
    kpx = cmath.sqrt(k_px**2)
    return (2*e*cmath.exp(kpx*d))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))  +  (e*cmath.exp(0.5 * kpx*d)*((e-1)*cmath.exp(2*d2*kpx) - (e+1)))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))
```

```python
def lam_dash_s_pos(v, k_px):
    e = complex(eps_r(v), eps_i(v))
    kpx = cmath.sqrt(k_px**2)
    return (k_px/abs(k_px)) * (-(2*e)/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))  +  (e*cmath.exp(-0.5 * kpx*d)*((e-1)*cmath.exp(2*d2*kpx) - (e+1)))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d))))

def lam_dash_s_neg(v, k_px):
    e = complex(eps_r(v), eps_i(v))
    kpx = cmath.sqrt(k_px**2)
    return (k_px/abs(k_px)) * ((2*e*cmath.exp(kpx*d))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d)))  -  (e*cmath.exp(0.5 * kpx*d)*((e-1)*cmath.exp(2*d2*kpx) - (e+1)))/(k_px*((e-1)*((d1+d2)*cmath.exp(kpx*(d1+d2)) - d3*cmath.exp(kpx*d3)) + (e+1)*d*cmath.exp(kpx*d))))
```

```python
# Function for real part of electric conductivity

def sigma_er(v, k_px):
    eps = complex(eps_r(v), eps_i(v))
    lam_p = lam_dash_a_pos(v, k_px)
    lam_n = lam_dash_a_neg(v, k_px)
    kpx = cmath.sqrt(k_px**2)

    return (-k0*T/2)*(abs(kpx)**2)*eps.imag/(abs(eps)**2 * k_px.imag) * ((lam_p.real**2 + lam_p.imag**2)/(-2 * kpx.real) * (math.exp(2*kpx.real*d2) - math.exp(2*kpx.real*(d2+d3))) + (lam_n.real**2 + lam_n.imag**2)/(2 * kpx.real) * (math.exp(-2*kpx.real*d2) - math.exp(-2*kpx.real*(d2+d3))))
# Function for imaginary part of electric conductivity

def sigma_ei(v, kpx_arr):
    step = 1.0/100
    freq_range = np.arange(step, 1 + step, step)
    
    sum = 0.0
    
    for i in range(100):
        if freq_range[i] != v:
            integrand = -sigma_er(freq_range[i], kpx_arr[i])/(freq_range[i] - v)
            if not np.isnan(integrand):
                sum+=integrand
    return sum/314.0
```

```python
# Function for real part of magnetic conductivity

def sigma_mr(v, k_px):
    eps = complex(eps_r(v), eps_i(v))
    lam_p = lam_dash_s_pos(v, k_px)
    lam_n = lam_dash_s_neg(v, k_px)
    kpx = cmath.sqrt(k_px**2)
    return (k0*T/2)*(abs(kpx)**2)*eps.imag/(abs(eps) * k_px.imag) * ((lam_p.real**2 + lam_p.imag**2)/(-2 * kpx.real) * (math.exp(2*kpx.real*d2) - math.exp(2*kpx.real*(d2+d3))) + (lam_n.real**2 + lam_n.imag**2)/(2 * kpx.real) * (math.exp(-2*kpx.real*d2) - math.exp(-2*kpx.real*(d2+d3))))
```

```python
# Function for imaginary part of magnetic conductivity

def sigma_mi(v, kpx_arr, theta):
    step = 1.0/100
    freq_range = np.arange(step, 1 + step, step)
    
    sum = 0.0
    
    for i in range(100):
        if freq_range[i] != v:
            integrand = np.sin(theta)**2 * sigma_mr(freq_range[i], kpx_arr[i])/(freq_range[i] - v)
            if not np.isnan(integrand):
                sum+=integrand
    return sum/314.0
```

```python
# Function to calculate reflection coefficient

def reflection_coeff(v, sig_e, sig_m, theta):
    sig_e = sig_e/z0
    sig_m = sig_m*z0
    eps = complex(eps_r(v), eps_i(v))
    numerator = (-4 * eps * sig_m + 4 * sig_e * z0**2 * np.cos(theta) * np.sqrt(eps - np.sin(theta)**2) - sig_e * sig_m * z0 * np.sqrt(eps - np.sin(theta)**2) + eps * sig_e * sig_m * z0 * np.cos(theta) - 4 * z0 * np.sqrt(eps - np.sin(theta)**2) + 4 * eps * z0 * np.cos(theta))
    denominator = (4 * eps * sig_m + 4 * sig_e * z0**2 * np.cos(theta) * np.sqrt(eps - np.sin(theta)**2) + sig_e * sig_m * z0 * np.sqrt(eps - np.sin(theta)**2) + eps * sig_e * sig_m * z0 * np.cos(theta) + 4 * z0 * np.sqrt(eps - np.sin(theta)**2) + 4 * eps * z0 * np.cos(theta))
    
    return numerator / denominator
```

```python
# Function for real part of electric conductivity

def sigma_er_blunt(v, k_px, phi):
    eps = complex(eps_r(v), eps_i(v))
    lam_p = lam_dash_a_pos(v, k_px)
    lam_n = lam_dash_a_neg(v, k_px)
    kpx = cmath.sqrt(k_px**2)
    L = d
    term1 = 7*k0 * T / 2 * abs(k_px)**2 * eps.imag / abs(eps)**2
    term2_denom = abs(1 - cmath.exp(1j * k_px * L + 1j * phi))**2
    term2 = ((abs(lam_p)**2 / (-2 * kpx.real) *
              (cmath.exp(2 * kpx.real * d2) - cmath.exp(2 * kpx.real * (d2 + d3))))
             + (abs(lam_n)**2 / (2 * kpx.real) *
                (cmath.exp(-2 * kpx.real) * d2) - cmath.exp(-2 * kpx.real * (d2 + d3)))) * (1 - cmath.exp(-2 * k_px.imag * L)) / k_px.imag - (np.conjugate(lam_p) * (lam_n) /
                (1j * 2 * k_px.imag) *
                (cmath.exp(-1j * 2 * kpx.imag * d2) - cmath.exp(-1j * 2 * kpx.imag * (d2 + d3)))
             + (lam_p * np.conjugate(lam_n) /
                (-1j * 2 * k_px.imag)) *
             (cmath.exp(1j * 2 * kpx.imag * d2) - cmath.exp(1j * 2 * kpx.imag * (d2 + d3))) *
             (2 * cmath.exp(k_px.imag * L) / k_px.real) *
             (np.sin(k_px.real * L + phi) - np.sin(phi)))
    # Combine terms to get the final result
    sigma_er = term1 / term2_denom * term2
    
    return sigma_er
```

## 9. Breaking the Symmetry

In this section, we discuss how both the symmetric and anti-symmetric bands can be excited by breaking the symmetry. Two ways for breaking the symmetry will be considered -


*   Symmetric metasurface under oblique incidence - source breaks the symmetry and both bands can be excited
*   Asymmetric metasurface - asymmetry in the geometry mixes the antisymmetric and symmetric modes so that they become coupled with each other






### Symmetric metasurface under oblique incidence

When the incident field is coming at an oblique angle of incidence, the source is no longer symmetric with respect to the y = 0 plane. As a consequence, the dark mode becomes bright, and both the antisymmetric and the symmetric bands are excited. Since both antisymmetric and symmetric modes are excited, there will be both an electric surface conductivity $\sigma_{e}$ and a
 magnetic surface conductivity $\sigma_{m}$.

For the groove metasurface, the lower band ($\omega_{c1} \lt \omega \lt \omega_{sp}$) is the antisymmetric mode, and its energy dissipation is modeled as a complex electric surface conductivity $\sigma_{e}$. Since the conductivity only depends on the intrinsic absorption cross section of SPPs and not on the incidence angle, it is the same as for the normal indidence case considered before. On the other hand, the upper band is the symmetric mode ($\omega_{sp} \lt \omega \lt \omega_{c1}$) and gives magnetic surface conductivity $\sigma_{m}$. Finally, for reflectivity we get two continuous bands corresponding to the excitation of the antisymmetric and symmetric modes instead of just one.

```python
# Function for real part of magnetic conductivity

def sigma_mr_blunt_ob(v, k_px, phi):
    eps = complex(eps_r(v), eps_i(v))
    lam_p = lam_dash_a_pos(v, k_px)
    lam_n = lam_dash_a_neg(v, k_px)
    kpx = cmath.sqrt(k_px**2)
    L = d
    term1 = -np.sin(0.4*np.pi)**2 * k0 * T / 2 * abs(k_px)**2 * eps.imag / abs(eps)**2
    term2_denom = abs(1 + cmath.exp(1j * k_px * L + 1j * phi))**2
    term2 = ((abs(lam_p)**2 / (-2 * kpx.real) *
              (cmath.exp(2 * kpx.real * d2) - cmath.exp(2 * kpx.real * (d2 + d3))))
             + (abs(lam_n)**2 / (2 * kpx.real) *
                (cmath.exp(-2 * kpx.real) * d2) - cmath.exp(-2 * kpx.real * (d2 + d3)))) * (1 - cmath.exp(-2 * k_px.imag * L)) / k_px.imag - (np.conjugate(lam_p) * (lam_n) /
                (1j * 2 * k_px.imag) *
                (cmath.exp(-1j * 2 * kpx.imag * d2) - cmath.exp(-1j * 2 * kpx.imag * (d2 + d3)))
             + (lam_p * np.conjugate(lam_n) /
                (-1j * 2 * k_px.imag)) *
             (cmath.exp(1j * 2 * kpx.imag * d2) - cmath.exp(1j * 2 * kpx.imag * (d2 + d3))) *
             (2 * cmath.exp(k_px.imag * L) / k_px.real) *
             (np.sin(k_px.real * L + phi) - np.sin(phi)))
    # Combine terms to get the final result
    sigma_mr = term1 / term2_denom * term2
    
    return sigma_mr
```

```python
# Function for real part of electric conductivity

def sigma_er_blunt_asymm(v, k_px, phi):
    eps = complex(eps_r(v), eps_i(v))
    lam_p = lam_dash_a_pos(v, k_px)
    lam_n = lam_dash_a_neg(v, k_px)
    kpx = cmath.sqrt(k_px**2)
    L = d
    term1 = 4.5*k0 * T / 2 * abs(k_px)**2 * eps.imag / abs(eps)**2
    term2_denom = abs(1 - cmath.exp(1j * k_px * L + 1j * phi))**2
    term2 = ((abs(lam_p)**2 / (-2 * kpx.real) *
              (cmath.exp(2 * kpx.real * d2) - cmath.exp(2 * kpx.real * (d2 + d3))))
             + (abs(lam_n)**2 / (2 * kpx.real) *
                (cmath.exp(-2 * kpx.real) * d2) - cmath.exp(-2 * kpx.real * (d2 + d3)))) * (1 - cmath.exp(-2 * k_px.imag * L)) / k_px.imag - (np.conjugate(lam_p) * (lam_n) /
                (1j * 2 * k_px.imag) *
                (cmath.exp(-1j * 2 * kpx.imag * d2) - cmath.exp(-1j * 2 * kpx.imag * (d2 + d3)))
             + (lam_p * np.conjugate(lam_n) /
                (-1j * 2 * k_px.imag)) *
             (cmath.exp(1j * 2 * kpx.imag * d2) - cmath.exp(1j * 2 * kpx.imag * (d2 + d3))) *
             (2 * cmath.exp(k_px.imag * L) / k_px.real) *
             (np.sin(k_px.real * L + phi) - np.sin(phi)))
    # Combine terms to get the final result
    sigma_er = term1 / term2_denom * term2
    
    return sigma_er
```

```python
# Function for imaginary part of electric conductivity

def sigma_ei_blunt_asymm(v, kpx_arr, phi):
    step = 1.0/100
    freq_range = np.arange(step, 1 + step, step)
    
    sum = 0.0
    
    for i in range(100):
        if freq_range[i] != v and kpx_arr[i].real !=0 and kpx_arr[i].imag !=0:
            integrand = -sigma_er_blunt(freq_range[i], kpx_arr[i], phi[i])/(freq_range[i] - v)
            if not np.isnan(integrand):
                sum+=integrand
    return sum/314.0
```

## Conclusion
The analytical theory presented herein provides a comprehensive examination of the optical response exhibited by singular plasmonic metasurfaces. Through the application of transformation optics, it has been demonstrated that subwavelength metasurfaces featuring singular grooves or wedges exhibit spectral equivalence to a more symmetric system, specifically a simple periodic array of metal slabs. This equivalence facilitated the derivation of analytical expressions for the dispersion relation of surface plasmon polariton (SPP) modes within the metasurface, as well as for the absorption cross section. Furthermore, it has been elucidated that under normal incidence, only the anti-symmetric band is excited, positioned below the surface plasmon frequency for grooves (and above for wedges). Introducing asymmetry through oblique angle incidence allows for the excitation of the symmetric band, positioned above the surface plasmon frequency (and below for wedges).

Additionally, the incorporation of a flat surface model has enabled the derivation of effective surface conductivities to characterize energy dissipation by SPPs, leading to analytical expressions for metasurface reflectivity. Notably, it has been demonstrated that singular plasmonic metasurfaces exhibit continuous spectra, resulting in reduced reflectivity across a broad range of frequencies. However, the practical realization of perfect singularities is hindered by challenges in nanofabricating sharp angles and the eventual onset of non-local effects. Blunt singularities, representing a departure from idealized singularities, have been shown to significantly impact the spectrum of metasurfaces. Specifically, blunt singular metasurfaces map to truncated slab arrays, resulting in the quantization of SPP modes and the emergence of a discrete set of peaks in the spectrum. Finally, the excitation of both anti-symmetric and symmetric bands at normal incidence in asymmetric metasurfaces has also been discussed.
